```python
import asyncio
import ssl
import certifi
import websockets

ssl_context = ssl.create_default_context(
    cafile=certifi.where()
)

async def main():
    uri = "wss://stream.binance.com:9443/ws/btcusdt@aggTrade"
    async with websockets.connect(uri, ssl=ssl_context) as ws:
        while True:
            print(await ws.recv())

await main()
```

```python
import asyncio
from collections import defaultdict

# Shared in-memory data
user_subscriptions = defaultdict(set)
symbol_subscribers = defaultdict(set)

async def subscribe(user_id, symbols):
    for symbol in symbols:
        # "Synchronous" operations
        user_subscriptions[user_id].add(symbol)
        symbol_subscribers[symbol].add(user_id)

        # Yield control to the event loop
        await asyncio.sleep(0)

    print(f"{user_id} done: {user_subscriptions[user_id]}")

async def main():
    symbols = ["BTC", "ETH"]
    
    # Run two subscriptions concurrently
    await asyncio.gather(
        subscribe("user1", symbols),
        subscribe("user2", symbols)
    )

    print("Final state:", dict(symbol_subscribers))

await main()
```
